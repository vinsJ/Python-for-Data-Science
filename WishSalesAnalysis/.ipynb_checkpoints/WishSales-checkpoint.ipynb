## 1.3 Cleaning up the DataSet

Here, we will try and make our dataset more "analysable". We'll need to decide **wich columns to keep and wich columns to get ride of** based on how we want to analyse it.

The main goal of this analysis is to check and see if we find some correlation between the retail price and price (If people get baited by the urge of buying someting with a *"discount"* and basicaly see what can **influences sales of an item**.

```python
salesAllCol[salesAllCol['currency_buyer'] != 'EUR']
```

We've checked that all the items are in Euros, the **curency_buyer** column is therefore, meaningless

```python
sales = salesAllCol[['title','price','retail_price','units_sold','rating','uses_ad_boosts','product_color','countries_shipped_to','shipping_option_price','has_urgency_banner','merchant_rating_count','merchant_rating']]
```

It could be interesting to have a column showing the **discount** shown by Wish on an item. ***Important :*** When an item has a negative discount, there are no discount shown on the website. (For instance: https://www.wish.com/product/2020-summer-vintage-flamingo-print--pajamas-set-casual-loose-t-shirt-top-and-elastic-shorts-women-sleepwear-night-wear-loungewear-sets-5e9ae51d43d6a96e303acdb0?hide_login_modal=true&share=web)

```python
price = (sales['retail_price'] - sales['price'])
totalPrice = (sales['retail_price'] + sales['shipping_option_price'])
```

```python
sales.insert(3,"discount",price)
sales.insert(8,'totalPrice',totalPrice)
```

Now that we have **meaningfull data**, let's analyse our data

```python
sales.columns
```

## 2.3 Plot questions

Now that we've seen a lot of information, let's try and analyse the data using **graphs**.

### 2.3.1 Does the quality have a influence over the price ? 

```python
plt.scatter(sales.rating,sales.price)

plt.xlabel('Rating')
plt.ylabel('Price')
```

Hum, looks like there is no correlation ***at all*** ! 

Let's try to group by the rating and show the mean of the price

```python
sales.groupby('rating')['price'].mean().plot()
```

Again, we can see that the quality of an item does not influence its price.

Let's do the same thing on the items sold 100 times. 

```python
sales[(sales.units_sold == 1000)].groupby('rating')['price'].mean().plot()
```

### 2.3.2 What does influence the number of sales ? 

We could think that the greatest the discount is, better the sales are. Let's check it out ! 

```python
plt.plot(sales.groupby('units_sold')['discount'].mean(),color='red',label='discount')

plt.plot(sales.groupby('units_sold')['price'].mean(),color='blue',label='price')

plt.plot(sales.groupby('units_sold')['totalPrice'].mean(),color='orange',label='totalPrice')

plt.yscale("log")
```

I've used a **logarithmic scale** in order to have a better view on the variations. 

It seems that the ***price*** have some influence over the ***units sold***.

```python
plt.plot(sales.groupby('units_sold')['rating'].mean(),color='green',label='rating')

plt.plot(sales.groupby('units_sold')['merchant_rating'].mean(),color='grey',label='merchant_rating')

plt.yscale("log")
```

### 2.3.3 Last but not least, does the merchant rating influence the price ?

```python
plt.plot(sales.groupby('units_sold')['merchant_rating'].mean(),color='red')

plt.yscale("log")
```

Well, the merchant rating seems not to have much influence on the number of sales.
