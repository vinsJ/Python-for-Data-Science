## Boys/Girls disparity

You must ask yourself, what is **boys/girls** disparity ? 

To do that, let's print a choropleth map using ratio of boys/girls.

We don't need weighted by the total population of the dpt because this is a ratio, however if we used boys - girls, we'd need to weighted by the total births in the dpt. 

*If we use just boys-girls, our results would be meaningless because, we cannot compare boys-girls in a dpt of 20 people and a dpt of 100k people !*

**Enough talk ! Let's dive into it :)**

```python
dpt_sexe = prenoms.groupby(['dpt','sexe']).nombre.sum()
dpt_sexe = dpt_sexe.reset_index()
dpt_sexe
```

Now, we've got our data frame grouped by dpt and sex, we juste need to calculate the ratio and then merge the dataframes

```python
def get_ratio(df):
  ratios = list()
  for i in range(0,len(df)-1,2):
    ratio = df['nombre'][i]/df['nombre'][i+1]
    ratios.append(ratio)
  return ratios
```

```python
dpt_ratios = prenoms.groupby('dpt').nombre.sum()
dpt_ratios = dpt_ratios.reset_index()
dpt_ratios.insert(1,'ratio', get_ratio(dpt_sexe), True)
dpt_ratios = dpt_ratios[['dpt','ratio']]
dpt_ratios
```

Let's merge this DataFrame to the other one ! 

```python
dpt_ratios["code_insee"] = dpt_ratios.dpt
dpt_ratios
```

```python
merged_ratios = map_df.merge(dpt_ratios,on='code_insee')
merged_ratios
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, figsize=(8, 6))

merged_ratios.plot(column="ratio",   ax=ax, linewidth=0.3,edgecolor="0.8",legend=True,legend_kwds
            ={"label": "Ratios of boys/girls births"})
```

**How to interpret the ratio and the map ?**

We calculated numberBoys/numberGirls, so :

ratio = 1 => same number of boys and girls

ratio < 1 => more girls than boys

ratio > 1 => more boys than girls

### Most innovant departements 

How can we define Innovation ? 

That could be the rate of new names on a given year, or the avg of rates on x years. 

```python
# Some tests I did performing the calcul of unique firstnames in a dpt during a certain year

prenomsY = prenoms.groupby(['annais','dpt']).agg({"preusuel": "unique"})
prenomsY = prenomsY.reset_index()
prenomsY
```

```python
def ListeUniquePrenom(df, anneeListe):
    liste = list()
    dfSliced =  df[df.annais < anneeListe]
    for rows in dfSliced['preusuel']:
        liste.extend(rows)

    mySet = set(liste)
    liste = list(mySet)
    return liste    
```

```python
# print(ListeUniquePrenom(prenomsY,2017))

def nbNewPrenoms(listePrenomsAll, df, anneeInnov):
    listeNbNewPrenoms = list()
    dFI = df[df.annais == anneeInnov]
    for i in range(len(dFI)):
        dep = dFI.iloc[i]['dpt']
        listeP = dFI.iloc[i]['preusuel']
        cpt = 0
        for j in range(len(listeP)):
            if(listeP[j] not in listePrenomsAll):
                cpt += 1
        listeNbNewPrenoms.append(cpt)
    return listeNbNewPrenoms
```

```python
listePre = ListeUniquePrenom(prenomsY,'2017')
listeNbNewPrenoms = nbNewPrenoms(listePre,prenomsY,'2018')

prenomsY['nbNewPrenom'] = listeNbNewPrenoms
prenomsY
```
